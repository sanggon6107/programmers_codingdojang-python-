```python
print("="*50, "2019.08.16", "="*50)

"""""""""""""""""""""""""""""

import sys
from PyQt5.QtWidgets import QApplication, QWidget, QPushButton, QLabel #메인 윈도우(initUI) 모듈
from PyQt5.QtWidgets import QMainWindow #스테이터스바를 위한 모듈
from PyQt5.QtWidgets import QHBoxLayout, QVBoxLayout # 박스 레이아웃 모듈(수평, 수직)
from PyQt5.QtCore import QCoreApplication #버튼 이벤트(quit) 모듈
from PyQt5.QtCore import QDateTime # DATETIME 모듈


class MyApp1(QWidget):

    def __init__(self):
        super().__init__()

        self.initUI()
        
    def initUI(self):
        
        button1 = QPushButton('button1')
        button1.clicked.connect(QCoreApplication.instance().quit)
        button2 = QPushButton('button2')
        button2.clicked.connect(QCoreApplication.instance().quit)
        
        PyQt5.DATA_1  == connect(overflow(EXIT)):
            import(DATA_1) = label('DATA1 :'):
                for N in [121. 31. 55. 814]:
                    if type(N) != int:
                        break;
                
                    elif N => 31:
                        continue;
                    else :
                        DATA_1 = CALL[N]
                        
                    N = N+1
                    
                    continue
                    
                
                
    def imp_DATA(a, b):
        
        import re
        
        word_RE = re.find('QCoreApplication')
        
        Found_D = re.search(word_RE, 'a')
        
        if not Found_D :
            print("객체를 찾을 수 없습니다.")
        
        if Found_D :
            print(Found_D)
            
        
        
        hbox = QHBoxLayout()    #수평 박스레이아웃
        hbox.addStretch(1)
        hbox.addWidget(button1)
        hbox.addWidget(button2)
        hbox.addStretch(1)
        
        vbox = QVBoxLayout()    #수직박스레이아웃 : 수평박스레이아웃의 객체를 대입
        vbox.addStretch(1)
        vbox.addLayout(hbox)
        vbox.addStretch(1)
        
        ## 'self.메서드'로 표현된 코드는  MyApp1클래스에서 상속한 QWidget클래스 안에 포함된 메서드.
        ## 따라서, 코드 앞 쪽에서 클래스나 메서드 이름을 호출할 필요가 없다.
        
        self.setLayout(vbox)       
        
        self.setWindowTitle('PyQt5 Layout')
        self.setGeometry(700, 300, 300, 200)
        self.show()
        
if __name__ == '__main__':

    app = QApplication(sys.argv)
    ex = MyApp1()
    sys.exit(app.exec_())

"""""""""""""""""""""""""""""""""""""""""

import sys
from PyQt5.QtWidgets import (QApplication, QWidget, QGridLayout, QLabel, QLineEdit, QTextEdit)
from PyQt5.QtCore import QDateTime # DT데이터 모듈
from PyQt5.QtWidgets import QMainWindow # 스테이터스바 모듈

class MyMainWindow(QMainWindow):
    def __init__(self):
        super().__init__()
        self.main_widget = MyApp(self)   # MyApp 클래스를 호출하는 객체를 선언
        self.setCentralWidget(self.main_widget) 
        # MyMainWindow클래스에 존재하는 setCentralWidget메서드를 호출하고, 변수에 self.main_widget오브젝트를 대입한다.

        ## self.main_widget을 반드시 선언하여야 하는가?
        ## self.main_widget을 선언하지 않고, self.setCentralWidget 메서드에 직접 MyApp(self)클래스를 대입해도 된다.
        
        self.title = QLabel('Title :')
        self.author = QLabel('Author :')
        self.review = QLabel('Review :')
        
        
        self.initUI()


    def initUI(self):
        
        QDT = QDateTime.currentDateTime()
        self.statusBar().showMessage(QDT.toString('yyyy년 MM월 dd일'))
        
        self.setWindowTitle('QGridLayout')
        self.setGeometry(500, 500, 300, 200)
        self.show()


class MyApp(QWidget):

    def __init__(self, parent):
        super(MyApp, self).__init__(parent)
        self.parent = parent
        self.initUI()
        
        
        
    def initUI(self):

        grid = QGridLayout()
        self.setLayout(grid)
        
        

        grid.addWidget(self.title, 0, 0)
        grid.addWidget(self.author, 1, 0)
        grid.addWidget(self.review, 2, 0)

        grid.addWidget(QLineEdit(), 0, 1)
        grid.addWidget(QLineEdit(), 1, 1)
        grid.addWidget(QTextEdit(), 2, 1)


        self.setWindowTitle('QGridLayout')
        self.setGeometry(500, 500, 300, 200)
        self.show()
        



if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyMainWindow()
    sys.exit(app.exec_())
```
