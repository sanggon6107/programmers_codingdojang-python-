```python
print("Getting GEKKO optimal control solver started")



##이번 Numerical solution 에서는 Lagrange Problem을 Mayer Problem으로 변환한 후 수치계산 

from gekko import GEKKO
import numpy as np
import matplotlib.pyplot as plt  

m = GEKKO() # initialize gekko
nt = 301
m.time = np.linspace(0,30,nt)

#constant
gamma = 1.0


# Variables
x1 = m.Var(value = 1)
x2 = m.Var(value = 0) 
x3 = m.Var(value = 0)#Lagrange_Problem -> Mayer_Problem 
u = m.MV(value = 0, lb=-7,ub=7)
u.STATUS = 1

p = np.zeros(nt) # mark final time point
p[-1] = 1.0
final = m.Param(value=p)

# Equations
m.Equation(x1.dt()== x2)
m.Equation(x2.dt()== -1*x1 + u)
m.Equation(x3.dt()== (1/2)*((x1)**2 + (gamma)*(u)**2))

m.Obj(x3*final) # Objective function

m.options.IMODE = 6 # optimal control mode
m.options.NODES = 3 # collocation nodes
m.options.SOLVER = 3 # solver (IPOPT)



m.solve(disp=True) # solve


plt.figure(1, figsize = (11, 5)) # plot results
plt.style.use('ggplot')
plt.plot(m.time,x1.value,'k-',label=r'$x_1$')
plt.plot(m.time, x2.value, 'b-', label = r'$x_2$')
plt.plot(m.time,u.value,'r--',label=r'$u$')

plt.legend(loc='best')
plt.xlabel('Time[sec]')
plt.ylabel('Value')
plt.show()


print(u.value)
```
