```python
import re

s = '103450000'
zero_com = re.compile('[0]')
def in_col(string) :
    queue_1 = [string]
    #exi_nums = {queue_1[0]}
    #print(exi_nums)
    while '0' in queue_1[0] : #행 string에 대해, 전체 탐색 시행. !!!이전 탐색 결과와 중복(열방향 확인)은 제거하지 않음.!!!
        
        p_data = queue_1.pop(0)
        exi_nums = set(list(p_data))
        for alt in range(1, 10) :
            if not str(alt) in exi_nums :
                pp = zero_com.sub(str(alt), p_data, count = 1)
                queue_1.append(pp)
    print(queue_1)
    print(len(queue_1))


in_col(s)
```
