```python
def fib(n) :
    if n == 1 :
        return 1
    elif n == 2 :
        return 2
    else :
        return fib(n-1)+fib(n-2)
print(fib(40))
```

```python
def Dynamic_Programming(n, li) :
    if n == 1  or n == 2 :
        return li[n-1]
    else :
        if li[n-1] != '#' : return li[n-1]
        elif li[n-1] == '#' :
            li[n-1] = Dynamic_Programming(n-1, li)+ Dynamic_Programming(n-2, li)
    return li[n-1]

def main(N, sec) :
    LI = [1]+[sec]+['#' for i in range(N-2)]
    return Dynamic_Programming(N, LI)
print(main(1000, 2))
```

```python
def addWithoutPlus(a, b):
    if b == 0:
        return a
    elif a == 0:
        return b
    _sum = a^b
    _carry = (a&b) << 1
    return addWithoutPlus(_sum, _carry)

print(addWithoutPlus(100, 23))
```
