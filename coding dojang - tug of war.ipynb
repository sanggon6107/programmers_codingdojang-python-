```python
### 솔트 -> 스왑의 형태로 설계할 것.
def inp() :
    numb_p, w_list = int(input("N : ")), []
    for i in range(numb_p) :
        w_list.append(int(input("INPUT : ")))
        w_list_1 = w_list[:int(len(w_list)/2)]
        w_list_2 = list(set(w_list)-set(w_list_1)) # 집합 연산이므로 리스트의 index값 변동성 있음

    return numb_p, w_list_1, w_list_2

'''
subject to : abs(len(w_L_1)-len(w_L_2)) <= 1
minimize : abs(w_L_1_D)
'''

###  최초 sorting : 통상 소트 후 단순 분리할 것인가? sorting 후 교차로 나눌 것인가?
'''
 optimal solution 찾을 떄 까지 무한 루프(변분 = 0, 즉 ideal한 optimal solution을 찾지 못하더라도, 인접 상태량의 변화에 의한
 목적함수의 값이 일정하다면, 최소한 local optimal solution일테고, 선형 계획법의 특성상 곧 global optimal solution일 가능성이 크다.)
(KKT조건, 최적성의 2차 필요 조건)

mat과 w_L_1_D의 값 비교, 부호가 같으면서 가장 가까운 값이 mat[a][b]일 때, w_L_1[a]와 w_L_2[b]의 값을 swap한다.

local optimal solution 판별법 고려할 것.

'''

def M_pr(N, w_L_1, w_L_2) :
    ave = (sum(w_L_1)+sum(w_L_2))/2
    w_L_1_D, w_L_2_D, mat = sum(w_L_1)-ave, sum(w_L_2)-ave, []
    
    while True : 
        for pos in w_L_1 :
            mat.append(list(map(lambda x : pos-x, w_L_2)))
        print(mat)
        
        # 실험용. 단순히 mat에 w_L_1_D(이상치)로부터의 편차만을 구한다.
        for pos_exp in range(0, len(mat)) :
            for A_pos_exp in range(0, len(mat[pos_exp])) :
                mat[pos_exp][A_pos_exp] -= w_L_1_D
        print("갱신값 : ", mat)  # 갱신값이 0에 가장 가까워야한다 : local 최적해에서 무한 루프 가능성
        
        break
        
    return ave, w_L_1_D, w_L_2_D

q, w, e = inp()
print(w, e)
print(M_pr(q, w, e))
```
