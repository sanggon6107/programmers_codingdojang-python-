```python
##### 총 단어수, -번쨰. 중복 회피 구현


import sys
import pyautogui
from PyQt5.QtWidgets import *
import csv
import random
import os

class MyApp(QWidget):

    def __init__(self):
        super().__init__()
        
        self.fresh_btn = QPushButton("갱신")
        self.seikai_btn = QPushButton("정답 확인")
        self.count = 0
        self.NumberLabel = QLabel("%d번째 단어" %self.count)
        self.The_Number_of_Words = QLabel("총 단어수 : 0")
        ##set label(1,0)~label(1,3)
        
        self.l10 = QLabel('-')
        self.l11 = QLabel('-')
        self.l12 = QLabel('-')
        self.l13 = QLabel('-')
        
        
        ##reading file_data
        
        self.file_data = open("D:word_file_data.csv", 'r')
        if not self.file_data :
            print("FILE DOES NOT EXIST.")
        
        self.header = next(self.file_data)
        self.lines_data = self.file_data.readlines()
        
        self.The_Number_of_Words.setText("총 단어수 : %d"%len(self.lines_data))
        
        if self.lines_data :
            for self.line in self.lines_data :
                print(self.line)
        
        self.file_data.close()
        
        ## file name
        
        self.filename = os.path.splitext("D:word_file_data.csv")[0]
        
        self.initUI()
        
        

    def initUI(self):
        
        grid = QGridLayout()
        self.setLayout(grid)
        
        
        
        grid.addWidget(QLabel("한국어 훈/음"), 0, 0)
        grid.addWidget(QLabel("한자"), 0, 1)
        grid.addWidget(QLabel("訓読み"), 0, 2)
        grid.addWidget(QLabel("音読み"), 0, 3)
        
        grid.addWidget(self.l10, 1, 0)
        grid.addWidget(self.l11, 1, 1)
        grid.addWidget(self.l12, 1, 2)
        grid.addWidget(self.l13, 1, 3)
        
        FILE_label = QLabel("FILE : ")
        FILE_NAME_label = QLabel(self.filename)
        
        
        FILE_NAME_label.setStyleSheet("coler: black;"
                                      "border-style: solid;"
                                      "border-width: 1px;"
                                      "border-color: #000000")
        
        import pyautogui
        
        pyautogui.click(x, y, clicks = 3, intervals = 0.1)
        
        
        grid.addWidget(self.NumberLabel, 2, 0)
        grid.addWidget(self.The_Number_of_Words, 2, 3)
        grid.addWidget(FILE_label, 3, 0)
        grid.addWidget(FILE_NAME_label, 3, 1)
        
        grid.addWidget(self.fresh_btn, 4, 1)
        grid.addWidget(self.seikai_btn, 4, 2)
        
        self.fresh_btn.clicked.connect(self.fresh_btn_connect)
        self.seikai_btn.clicked.connect(self.seikai_btn_connect)

        self.setWindowTitle('Words')
        self.setGeometry(300, 300, 300, 200)
        self.show()
        
        
    def fresh_btn_connect(self):
        self.count = self.count + 1
        
        self.randlen = len(self.lines_data) - 1
        state_data = random.randint(0, self.randlen)
        aa = self.lines_data[state_data]
        self.decre = aa.split(',')
        
        self.l10.setText("-")
        self.l11.setText(self.decre[1])
        self.l12.setText("-")
        self.l13.setText("-")
        
        self.self.lines_data[state_data] = "-"
        
        self.NumberLabel.setText("%d번째 단어" %self.count)
        
    def seikai_btn_connect(self):
        
        self.l10.setText(self.decre[0])
        self.l12.setText(self.decre[2])
        kel = len(self.decre[3])-1
        refined_decre3 = self.decre[3][:kel]
        self.l13.setText(refined_decre3)
        
if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```
