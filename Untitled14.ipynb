```python
print("="*50, "2019.09.05", "="*50)

import pyautogui
from PyQt5.QtWidgets import *
import sys
import time
import random

class MyApp(QWidget):

    def __init__(self):
        super().__init__()

        self.GUI_btn = QPushButton('Button')
        self.STOP_btn = QPushButton('STOP')
        self.initUI()

    def initUI(self):
        
        grid = QGridLayout()
        self.setLayout(grid)
        
        grid.addWidget(self.GUI_btn, 0, 1)
        grid.addWidget(self.STOP_btn, 0, 2)
        
        self.GUI_btn.clicked.connect(self.GUI_clicked)
        self.STOP_btn.clicked.connect(self.STOP_clicked)

        self.setWindowTitle('AutoP')
        self.setGeometry(300, 300, 300, 300)
        self.show()

    def GUI_clicked(self):
        while True :
            time.sleep(2)
            pyautogui.moveTo(random.randint(200, 800), random.randint(200, 800))
            print(pyautogui.position())
    def STOP_clicked(self):
        print("DONE")

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```
