```python
print("="*50, "2019.07.05", "="*50)

import numpy as np
import sys
from copy import copy

#Class, mothod

class Cal:
    def setdata(self, first, second) :
        self.first = first
        self.second = second

    def add(self):
        return self.first+self.second
    
    def sub(self):
        return self.first-self.second
    
    def div(self):
        return self.first/self.second
    
    def mul(self):
        return self.first*self.second
    
    
a = Cal()
a.setdata(4, 2)

print("(a.first, a.second) = (%d, %d)" % (a.first, a.second))
print("a.add() =", a.add())
print("a.sub() =", a.sub())
print("a.div() =", a.div())
print("a.mul() =", a.mul())

# class : constructor

class Cal2:
    def __init__(self, first, second) :
        self.first = first
        self.second = second

    def add(self):
        return self.first+self.second
    
    def sub(self):
        return self.first-self.second
    
    def div(self):
        return self.first/self.second
    
    def mul(self):
        return self.first*self.second
    
c = Cal2(5, 6)
print(c.first, c.second)


# class : Inheritance

class ForMoreCal(Cal2):
    def pow(self):
        return self.first ** self.second
    

d = ForMoreCal(6, 5)

print(d.pow())
print(d.add())

#class overwritting

#클래스 변수

class Family : 
    surname = "김"
    
a = Family()
a = Family.surname

print(a)
```
