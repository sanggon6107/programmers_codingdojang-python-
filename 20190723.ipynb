```python
print("="*50, "2019.07.23", "="*50)

import csv
from matplotlib import pyplot as plt

OPU = open("D:20190722103334.csv", 'r')

RI_U = csv.reader(OPU, delimiter = ",")
HEADER = next(RI_U)

extd_X_DATA = 0
X_DATA = []
Y_DATA = []

line_DATA = 0
for line_DATA in RI_U :
    print(line_DATA)     
    
    extd_X_DATA = line_DATA[1][5:]
    X_DATA.append(extd_X_DATA)
    Y_DATA.append(line_DATA[-2])
    
OPU.close()

f1 = plt.figure(1, figsize=(15, 10))
plt.subplot(1, 1, 1)
plt.plot(X_DATA, Y_DATA)
plt.title("The Average Temperature On June, 2018")
plt.xlabel("date")
plt.grid(True)
plt.ylabel("Temperature[C]")
plt.legend(['June(2018)'])
plt.show()


print(X_DATA)
print("\n\n", Y_DATA)
```
