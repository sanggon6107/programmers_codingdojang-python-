```python
import random
inp, li = input("input : "), ['ga', 'ba', 'bo', 'ga']
com = random.randint(0, 2)
print("COM : ", li[com])
if inp == li[com] :
    print("Draw")
elif inp == li[com+1] :
    print("You win")
else :
    print("You lose")
```
