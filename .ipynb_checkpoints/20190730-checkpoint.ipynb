```python
print("="*50, "2019.07.30", "="*50)

import csv
import re
from matplotlib import pyplot as plt


file_data = open("D:GEN_GLOBAL_FILE.csv", 'r')

read_data = csv.reader(file_data, delimiter = ",")

HEADER = next(read_data)

ip_data = input("지역 입력 : ")

line_data = 0

m = []
f = []

for line_data in read_data : 
    if ip_data in line_data[0] :
        for m_number in range(3, 104):
                m.append(-int(line_data[m_number]))
        for f_number in range(106, 207):
                f.append(int(line_data[f_number]))

                
print(m)
print("\n\n\n")
print(f)

file_data.close()

fig = plt.figure(1, figsize = (12, 10))
plt.barh(range(101), m, label = "MALE" )
plt.barh(range(101), f, label = "FEMALE")

plt.rc("font", family = "Malgun Gothic")
plt.style.use("ggplot")
plt.legend()

plt.show


print(HEADER)
```
