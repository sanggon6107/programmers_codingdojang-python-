```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pyautogui
import csv
import random

####################################
chrome_options = Options()
chrome_options.add_extension(r"C:\Users\user\Desktop\chrome extention(markdown) dir\1.1.6_0.crx")
#### extention중 크롤링에 사용될 기능을 crx화, driver의 옵션으로서 추가한다.
####################################
ch_dir = r'C:\Users\user\Desktop\ToLearnPy\chromedriver.exe'
driver = webdriver.Chrome(ch_dir,chrome_options = chrome_options)  ############  options = chrome_options 설정
driver.get("https://wikidocs.net/book/1896")
time.sleep(0.5)

file_open = open("wiki_list_1.csv", 'r')
read_DATA = list(csv.reader(file_open, delimiter = ','))
line_DATA = []

for line_DATA in read_DATA :
    print(line_DATA)

while True :
    To_click_btn = driver.find_element_by_link_text(read_DATA[random.randint(0, 240)][0])
    To_click_btn.click()
    time.sleep(5) 
    btn_position = pyautogui.locateOnScreen("btn_position.png")
    
    print(btn_position)
    
    pyautogui.click(btn_position)
    time.sleep(5)
    
    back_position = pyautogui.locateOnScreen("backspace.png")
    pyautogui.click(back_position)
```
