```python
dic = dict()

def gcd(a, b):
    global dic

    if (a, b) in dic: return dic[(a,b)]
    if b == 0:        return a

    ret = gcd(b, a % b)
    dic[(a, b)] = ret
    return ret

cnt = 0
for d in range(5, 12001):
    for n in range(d//3 + 1, d//2 + 1):
        cnt += gcd(d, n) == 1

print(cnt) # 7295372, 30.682 sec
```

```python
import math
from fractions import Fraction as fr

def gcd(a, b) :
    if b == 0 :
        return a
    else :
        return gcd(b, a % b)

def gys(y) :
    result = []
    for i in range(1, y+1) :
        if gcd(i, y) == 1 and fr(1, 2) > fr(i, y) > fr(1, 3) :
            result.append('#')
    return result.count('#')
sais = 0
for k in range(3, 12001) :
    sais += gys(k)
    if k % 1000 == 0 :
        print(k)
print(sais)
```
