```python
import re
def solution(arr):
    answer, arr_str = [], "".join(map(str, arr))
    com = re.compile('1+|2+|3+|4+|5+|6+|7+|8+|9+|0+')
    answer = com.findall(arr_str)
    return answer
solution([1,1,3,3,0,1,1])
```

```python
def solution(arr):
    answer = []
    for i in range(0, len(arr)) :
        if i == 0 or arr[i]!=arr[i-1] :
            answer.append(arr[i])
    return answer
solution([1,1,3,3,0,1,1])
```
