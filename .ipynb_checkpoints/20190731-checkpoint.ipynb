```python
print("="*50, "2019.07.31", "="*50)

import re
import csv
from matplotlib import pyplot as plt
import math

ip_data = input("입력 : ")

file_data = open("D:GEN_GLOBAL_FILE.csv", 'r')
read_data = csv.reader(file_data)
HEADER = next(read_data)

M_LIST = []
F_LIST = []
line_data = []
MF_NUMBER = 0
size = []

for line_data in read_data : 
    if ip_data in line_data[0] :
        for MF_NUMBER in range(3, 104) :
            M_LIST.append(int(line_data[MF_NUMBER]))
            F_LIST.append(int(line_data[MF_NUMBER+103]))
            size.append(math.sqrt(int(line_data[MF_NUMBER])+int(line_data[MF_NUMBER+103])))
        
file_data.close()    
#plt.scatter([], [])


f1 = plt.figure(1, figsize = (14, 10))

plt.scatter(M_LIST, F_LIST, s = size, c = range(101), cmap = 'jet', alpha = 0.6)
plt.plot(range(max(F_LIST)), range(max(F_LIST)), 'g')
plt.style.use('ggplot')
plt.title("plt.scatter()")
plt.xlabel("MALE")
plt.ylabel("FEMALE")
plt.colorbar()
plt.show
```
