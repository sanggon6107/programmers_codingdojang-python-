```python
from copy import deepcopy

r_data = '''0000000000
0000001000
0000110100
0011000010
0100000010
0100000010
0100000100
0010001000
0001011000
0000100000'''

data = [list(i) for i in  r_data.split()]

class painter :
    def __init__(self) :
        self.visited = []
    
    def BFS(self, map_, x, y) :
        queue = [map_+[[x, y]]]
        dx_dy = [(1, 0), (0, 1), (-1, 0), (0, -1)] # dx, dy : R, D, L, U
        
        while len(queue) <= 10000 :
            pop_data = deepcopy(queue.pop(0))
            #print(pop_data)
            for cond in dx_dy : 
                if (pop_data[pop_data[-1][1]+cond[1]][pop_data[-1][0]+cond[0]]) == '0' :#and (not pop_data[:-1] in self.visited) : #  
                    data_copy = deepcopy(pop_data)
                    data_copy[data_copy[-1][1]+cond[1]][data_copy[-1][0]+cond[0]] = '3'
                    if data_copy[:-1] in self.visited :
                        continue
                    data_copy[-1][1] += cond[1]
                    data_copy[-1][0] += cond[0]
                    
                    self.visited.append(data_copy[:-1])
                    queue.append(data_copy)
            print(len(self.visited))
            print(len(queue))
            self.printOut(queue)

        #return self.printOut(queue)
        
    def printOut(self, queue_) :
        for maps in queue_ :
            for line in maps[:-1] :
                print("".join(line))
            print("\n")
            break ###########################################################
        
        print("cycle")




if __name__ == '__main__' :
    a = painter()
    a.BFS(data, 5, 5)
```

```python
#   
# bfs  queue  .

from copy import deepcopy

r_data = '''0000000000
0000001000
0000110100
0011000010
0100000011
0100000000
0100000111
0010001000
0001011000
0000100000'''

data = [list(i) for i in  r_data.split()]

class painter :
    def BFS(self, map_, x, y) :
        queue = [[x, y]]
        dx_dy = [(1, 0), (0, 1), (-1, 0), (0, -1)]

        while len(queue) != 0 :
            pop_data = deepcopy(queue.pop(0))
            for cond in dx_dy :
                copy_data = deepcopy(pop_data)
                try :
                    if (map_[copy_data[1]+cond[1]][copy_data[0]+cond[0]]) == '0' :
                        map_[copy_data[1]+cond[1]][copy_data[0]+cond[0]] = '3'
                        queue.append([copy_data[0]+cond[0], copy_data[1]+cond[1]])
                except :
                    pass
        return self.printOut(map_)
        
    def printOut(self, map__) :
        for line in map__ :
            print("".join(line))

if __name__ == '__main__' :
    a = painter()
    a.BFS(data, 5, 5)
```
