```python
print("="*50, "2019.07.29", "="*50)

import csv
import re
from matplotlib import pyplot as plt

ip_data = input("지역 입력 : ")

f = open("D:GLOBAL_FILE.csv", 'r')

read_data = csv.reader(f, delimiter = ",")
HEADER_DATA = next(read_data)

line_data = 0
gra_data = []
n = 0

for line_data in read_data :
    if ip_data in line_data[0] :
        for n in line_data[3:] :
            gra_data.append(int(n))

    
f.close()

fig = plt.figure(1, figsize=(12, 10))
plt.bar(range(101), gra_data, )  ## 세로로 막대그래프를 그리고 싶다면 barth()
plt.style.use('ggplot')
plt.title("population by age")
plt.xlabel("age")
### plt.xticks(range(24), mx_station, rotation=90) x축 설정
plt.ylabel("population")
plt.show()

print("\n\n\n\n %s : " %ip_data , gra_data)

```
