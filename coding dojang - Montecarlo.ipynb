```python
import random
from matplotlib import pyplot as plt


x_list, y_list, x_list1, y_list1, count, x, y = [], [], [], [], 0, 0, 0

for n in range(int(input("N : "))) :
    x = random.uniform(0, 1)
    y = random.uniform(0, 1)
    
    if (x)**2+(y)**2 <= 1 :
        count +=1
        x_list.append(x)
        y_list.append(y)
    else :
        x_list1.append(x)
        y_list1.append(y)
print((count/n)*4)

f1 = plt.figure(1, figsize = (10, 10))
plt.subplot(1, 1, 1)
plt.scatter(x_list, y_list, c = 'blue')
plt.scatter(x_list1, y_list1, c = 'green')
plt.title("Montecarlo, 2019")
plt.grid(True)
plt.xlabel("x")
plt.ylabel("y")

plt.show()
```
