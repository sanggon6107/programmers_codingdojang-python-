```python
memoization = {1:1, 2:1}

def fib(n) :
    if n == 1 : return 1
    else :
        if n in memoization and not n-1 in memoization : return memoization[n], fib(n-1)
        if n in memoization and n-1 in memoization : return memoization[n], memoization[n-1]
        else :
            memoization[n] = fib(n-1)[0] + fib(n-2)[0]
            l = fib(n-1)
            return memoization[n], l

fib(80)
```
